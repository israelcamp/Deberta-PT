```python
from transformers import DebertaV2Tokenizer, AutoTokenizer
from transformers import PreTrainedTokenizerFast
from datasets import load_dataset, load_from_disk
import ftfy
from tqdm.auto import tqdm
```

```python
# dataset = load_dataset("wikipedia", language="pt", date="20230401",  beam_runner='DirectRunner')
dataset = load_from_disk("c4ai-wik")
```

```python
def batch_iterator(batch_size=10_000):
    for i in tqdm(range(0, len(dataset), batch_size)):
        text = dataset[i : i + batch_size]["text"]
        text = [ftfy.fix_text(t) for t in text]
        yield text
```

```python
tokenizer = AutoTokenizer.from_pretrained("deberta_pt_tokenizer")
```

```python
tokens = tokenizer.tokenize("Olá, mano, como que vai")
tokenizer.convert_tokens_to_string(tokens)
```
